Pie Chart

it is used when we have to show proportion

```python
sales_data = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
```

```python
sales_data
```

Now we have to Transpose the table and sum all values according to regions

```python
region_sales = sales_data.T.sum(axis='columns')
```

```python
region_sales
```

```python
plt.pie(region_sales, labels=region_sales.index,startangle=90, explode=(0.2,0,0,0), autopct='%1.1f%%')
```

by default in Matplotlib if i dont pass any bins in plt then by defualt 10 bins will be plot

bin means bars in histogram

```python
# in seaborn

sns.histplot(df['Year'], bins=10)
```

```python
# in seaborn we can turn our histogram in smooth curves which is known as KDE ( kernel density estimation)

sns.kdeplot(df['Year'])
```
